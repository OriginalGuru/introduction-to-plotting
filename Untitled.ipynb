```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
```

```python
k = 1.0 # the spring constant
x = np.linspace(-5, 5, 10) # define x as an np array
y = 0.5*k*x**2

xerr = np.random.rand(len(x)) # error in x
yerr = np.random.rand(len(x)) # error in y

# Adding labels, legend, and title
plt.xlabel('x-axis')
plt.ylabel('y-axis')

# np.transpose([x,y]) # testing the output of x and y
plt.plot(x, y, 
        marker=".", 
        linestyle="-", 
        linewidth=0.3, 
        color="#000000",
        markerfacecolor="red",
        markeredgecolor="blue")
plt.errorbar(x, y, xerr=xerr, yerr=yerr,
             fmt='.', 
             color='green',
             ecolor='black',
             elinewidth=0.3,
             capsize=3)

#plt.legend(loc='upper left')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(-10, 10, 100)
y = x**2

# Create the plot
fig, ax = plt.subplots()

# Plot the parabola
ax.plot(x, y)

# Move the x-axis and y-axis to the origin (0,0)
ax.spines['left'].set_position('zero')   # y-axis at x=0
ax.spines['bottom'].set_position('zero') # x-axis at y=0

# Hide the right and top spines (frame lines)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Set the ticks only on the bottom and left axes
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# Optional: Adjust the line width of the axes
ax.spines['left'].set_linewidth(0.1)
ax.spines['bottom'].set_linewidth(0.1)

# Show the plot
plt.show()
```
